## Importing Required Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("Libraries imported successfully")
```

## Accessing the Website

```python
url = "https://books.toscrape.com/"

response = requests.get(url)

print(response.status_code)
```

## Parsing HTML Content using BeautifulSoup

```python
soup = BeautifulSoup(response.text, "html.parser")

print(soup.title)
```

## Finding Book Containers from the Webpage

```python
books = soup.find_all("article")

print(len(books))
```

## Inspecting HTML structure of One Book

```python
print(books[0])
```

## Extracting Book Titles

```python
book_titles = []

for book in books:
    title = book.h3.a["title"]
    book_titles.append(title)

print(book_titles)
```

## Exracting Book Prices

```python
book_prices = []

for book in books:
    price = book.find("p", class_="price_color").text
    price = price.replace("Â", "")
    book_prices.append(price)

print(book_prices)
```

## Extracting Book Ratings

```python
book_ratings = []

for book in books:
    rating = book.find("p")["class"][1]
    book_ratings.append(rating)

print(book_ratings)
```

## Creating Structured Dataset using Pandas

```python
df = pd.DataFrame({
    "Book Title": book_titles,
    "Price": book_prices,
    "Rating": book_ratings
})

print(df)
```

## Exporting Data to CSV File

```python
df.to_csv("books_data.csv", index=False, encoding="utf-8-sig")

print("CSV file saved successfully")
```
